```python
from crewai import Agent,Task,Crew
from langchain_community.chat_models import ChatLiteLLM
api_key = "AIzaSyC9OYvhLULWHC_rE2fnYliIicc12IVwN5Y"

llm = ChatLiteLLM(
    model = "gemini/gemini-1.5-flash",
    temperature = 0.3,
    api_key = api_key
)
```

```python
match_details = """
ENG vs IND 2nd Test Report
The second Test between England and India, held at Edgbaston in Birmingham from July 2 to July 6, 2025, concluded with a resounding 336-run victory for India, leveling the series 1-1 and marking their first-ever Test win at that venue.
 This match was defined by standout performances from Shubman Gill and Akash Deep, who played pivotal roles in India’s dominance.

Batting Highlights
Shubman Gill, captaining India in this Test, delivered a record-breaking performance with the bat. He scored 269 in India's first innings and followed it up with 161 in the second innings, amassing 430 runs in the match. His leadership and batting prowess earned him the Player of the Match award.
 Other notable contributions came from Ravindra Jadeja, who scored 89 in the first innings, and Yashasvi Jaiswal, who added 87.

England’s batting was led by Jamie Smith, who scored 184 in their first innings, and Harry Brook, who contributed 158. In their second innings, Smith again stood out with 88, but no other batter could provide substantial support.

Bowling Highlights
India’s bowling attack was relentless, with Mohammed Siraj taking 6 for 70 in England’s first innings. However, the standout performer was Akash Deep, who took 4 for 88 in the first innings and followed it up with a maiden five-wicket haul of 6 for 99 in the second innings, finishing with a match haul of 10 wickets. This made him only the second Indian bowler after Chetan Sharma in 1986 to achieve such figures in English conditions.

Match Flow
India batted first and posted a massive total of 587, with Gill and Jadeja forming the backbone of the innings. England responded with 407, thanks to Smith and Brook’s efforts, but were always behind in the contest. India declared their second innings at 427 for 6, setting England a daunting target of 608. England never came close to that target, collapsing to 271 in their second innings, with Akash Deep once again at the forefront of the Indian bowling attack.

Key Moments
Akash Deep’s Spell: On the final day, Akash Deep’s 6 for 99 sealed the match for India, with his ability to exploit the conditions and the cracks on the pitch proving decisive.
Gill’s Leadership: Shubman Gill not only led from the front with the bat but also took the final catch to seal the victory, fittingly off Akash Deep’s bowling.
England’s Struggles: Despite Jamie Smith’s valiant 88 in the second innings, England never looked in control, with their batting line-up crumbling under pressure.
This victory was a historic moment for India, not only for the emphatic margin of the win but also for the emergence of new heroes like Akash Deep and the leadership of Shubman Gill. England, on the other hand, will need to regroup quickly as they head into the next Test at Lord’s, where they will aim to bounce back in the series.
"""
```

```python
#Agent 1 : News extrator
evaluator = Agent(
    role = "Gather the information from the provided match_details",
    goal = "Take the important key movements",
    backstory = "Expert in cricket report analysis",
    llm = llm,
    verbose = True
)
```

```python
#Agent 2 : Statistical Analyst
statistical_analyze = Agent(
    role ="Make a statistical report from the provided evaluation",
    goal = "Take the important key movements",
    backstory = "Expert in cricket report analysis",
    llm = llm,
    verbose = True
)
```

```python
#Agent 3 : Fan Voice
fan_voice = Agent(
    role ="Make a fan voice from the provided evaluation",
    goal = "Take the important key movements",
    backstory = "Expert in cricket report analysis",
    llm = llm,
    verbose = True
)
```

```python
#Agent 4 : Match Report
match_report = Agent(
    role ="Make a match report for the given match details",
    goal = "Make a neat match report",
    backstory = "Expert in cricket match report making",
    llm = llm,
    verbose = True
)
```

```python
# Task 1 : News Extraction
news_extraction = Task(
    description = f"""
    Get the match report and extract the news and information from the provided match report.\n\n{match_details}
    """,
    agent = evaluator,
    expected_output = "Gather the information from the provided match_details"
)
```

```python
# Task 2 : Statistical Analysis
analyze = Task(
    description  = "Take the important key movements , star performer, man of the match and some important statistical reports from the extracted news",
    agent = statistical_analyze,
    expected_output= "Key movements and star performance players with reeports"
)
```

```python
#Task 3 : Fan Voice
voice = Task(
    description = "Give emotional commentry of the performance in match by fans point of view",
    agent = fan_voice,
    expected_output = "Emotional commentry of the performance"

)
```

```python
#Task 4 : Match Report
report = Task(
    description = "Make a neat match report",
    agent = match_report,
    expected_output = "Neat match report as a PDF file"
)
```

```python
crew = Crew(
    agents =[evaluator,statistical_analyze,fan_voice,match_report],
    tasks = [news_extraction,analyze,voice,report]
)
```

```python
result = crew.kickoff()
print("Match ANalysis\n")
print(result)
```
